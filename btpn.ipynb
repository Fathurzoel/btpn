```python
#data cleansing
to_drop1 = ['CLIENTNUM','Customer_Age','Gender','Educationid','Maritalid',
           'card_categoryid','Months_on_book','Total_Relationship_Count',
           'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy',
            'Total_Trans_Amt', 'Total_Trans_Ct', 'Avg_Utilization_Ratio']
```

```python
# group by income category and calculate sum of total revolving balance
denda_by_income = data.groupby("Income_Category").agg({"Total_Revolving_Bal" : "sum"})

# rename column to Denda
denda_by_income = denda_by_income.rename(columns={"Total_Revolving_Bal": "Denda"})

# display result
display(denda_by_income)
```

```python
denda_by_income.plot(marker="o")

plt.show()
```

```python
result = data.groupby('Income_Category').agg({'Income_Category':'count', 'Total_Revolving_Bal':'sum'})
result.columns = ['Total_Customer', 'Denda']

#result.reset_index(inplace=True)

print(result)
```

```python
fig = plt.figure()
ax = result[['Denda']].plot(figsize=(20,15), kind='bar', color="b")
plt.xticks(rotation=0)
ax2 = ax.twinx()
ax2.plot(ax.get_xticks(), result[['Total_Customer']], marker='o', color="k")
plt.title('Kategori Penghasilan berdasarkan jumlah customer dan denda')
plt.show()
```

```python
profit = data.groupby('Income_Category').agg({'Income_Category':'count'})
profit.columns = ['Total_Customer']

print(profit)
```

```python
# mengambil data dari hasil pengelompokan dan perhitungan jumlah customer
profit = data.groupby('Income_Category').agg({'Income_Category':'count'})
profit.columns = ['Total_Customer']

# membuat visualisasi data pie
profit.plot(kind='pie', y='Total_Customer', 
            startangle=90, autopct='%1.1f%%', legend=False)
plt.axis('equal')
plt.title('Jumlah Customer berdasarkan Kategori Penghasilan')
plt.show()
```
